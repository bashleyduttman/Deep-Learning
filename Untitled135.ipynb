```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG16,ResNet50,InceptionV3
```

```python
train_data=tf.keras.utils.image_dataset_from_directory(
    directory='/content/rice leaf diseases dataset',
    image_size=(224,224),
    shuffle=True)
```

```python
for image,label in train_data.take(1):
  plt.figure(figsize=(10,10))
  for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(image[i].numpy().astype('uint8'))
    plt.title(train_data.class_names[label[i]])
```

```python
#custom model
input=tf.keras.layers.Input(shape=(224,224,3))
x=tf.keras.layers.Rescaling(1./255)(input)
x=tf.keras.layers.Conv2D(64,3,activation='relu')(x)
x=tf.keras.layers.Conv2D(16,3,activation='relu')(x)
x=tf.keras.layers.MaxPooling2D(strides=[1,1])(x)
x=tf.keras.layers.Flatten()(x)
output=tf.keras.layers.Dense(3,activation='softmax')(x)

```

```python
model1=tf.keras.Model(input,output)
model1.compile(loss='SparseCategoricalCrossentropy',optimizer='adam',metrics=['accuracy'])
model1_history=model1.fit(train_data,epochs=10)
```

```python
#initializing all the pretrained model
resnet=ResNet50(weights='imagenet',include_top=False,input_shape=(224,224,3))
for layers in resnet.layers:
  layers.trainable=False
vgg=VGG16(weights='imagenet',include_top=False,input_shape=(224,224,3))
for layers in vgg.layers:
  layers.trainable=False
inception=InceptionV3(weights='imagenet',include_top=False,input_shape=(224,224,3))
for layers in inception.layers:
  layers.trainable=False
```

```python
#training resnet model
x=tf.keras.layers.Flatten()(resnet.output)
output=tf.keras.layers.Dense(3,activation='softmax')(x)
rn_model=tf.keras.Model(resnet.input,output)
rn_model.compile(loss='SparseCategoricalCrossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
#training inceptionv3 model
x=tf.keras.layers.Flatten()(inception.output)
output=tf.keras.layers.Dense(3,activation='softmax')(x)
inception_model=tf.keras.Model(inception.input,output)
inception_model.compile(loss='SparseCategoricalCrossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
#training VGG16 model
x=tf.keras.layers.Flatten()(vgg.output)
output=tf.keras.layers.Dense(3,activation='softmax')(x)
vgg_model=tf.keras.Model(vgg.input,output)
vgg_model.compile(loss='SparseCategoricalCrossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
plt.figure(figsize=(5,5))
plt.plot(model1_history.history['accuracy'])
plt.legend(['custom_model'])
plt.xlabel('EPOCHS')
plt.ylabel('ACCURACY')
```
